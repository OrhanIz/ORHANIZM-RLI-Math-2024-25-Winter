## 19. Limits of Sequences

1. Calculate:
   - $\displaystyle \lim_{n \to \infty} \frac{n^2 + 3n}{2 n^2 - 2n}$

   - $\displaystyle \lim_{n \to \infty} \frac{(2n+3)^3}{n^3-1}$

2. Prove using the squeeze theorem:
   - $\displaystyle\lim_{n \to \infty} \frac{\sin(n)}{n}$

4. Find the limit of the sequence:
   - $a_n = (1+\frac{1}{n})^n$


# Solutions to Limits of Sequences

## 1. Calculating Limits

### **First Limit:**
$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n}
$$

#### **Solution:**
- Divide both numerator and denominator by \( n^2 \):

$$
\frac{n^2 + 3n}{2n^2 - 2n} = \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}
$$

- As \( n \to \infty \), the terms \( \frac{3}{n} \) and \( \frac{2}{n} \) approach 0, so:

$$
\lim_{n \to \infty} \frac{1 + 0}{2 - 0} = \frac{1}{2}
$$

Thus, the limit is:

$$
\frac{1}{2}
$$

---

### **Second Limit:**
$$
\lim_{n \to \infty} \frac{(2n+3)^3}{n^3-1}
$$

#### **Solution:**
- Expand the numerator:

$$
(2n+3)^3 = 8n^3 + 12n^2 \cdot 3 + 6n \cdot 9 + 27 = 8n^3 + 36n^2 + 54n + 27
$$

- Divide both numerator and denominator by \( n^3 \):

$$
\frac{8n^3 + 36n^2 + 54n + 27}{n^3 - 1} = \frac{8 + \frac{36}{n} + \frac{54}{n^2} + \frac{27}{n^3}}{1 - \frac{1}{n^3}}
$$

- As \( n \to \infty \), all small fraction terms go to 0:

$$
\lim_{n \to \infty} \frac{8 + 0 + 0 + 0}{1 - 0} = 8
$$

Thus, the limit is:

$$
8
$$

---

## 2. Proof Using the Squeeze Theorem

### **Limit to Prove:**
$$
\lim_{n \to \infty} \frac{\sin(n)}{n}
$$

#### **Solution:**
- We know that for all \( n \):

$$
-1 \leq \sin(n) \leq 1
$$

- Divide the entire inequality by \( n \):

$$
\frac{-1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}
$$

- Since \( \lim_{n \to \infty} \frac{1}{n} = 0 \) and \( \lim_{n \to \infty} \frac{-1}{n} = 0 \), by the **Squeeze Theorem**:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0
$$

## 3. Finding the Limit of the Sequence

### **Given Sequence:**
$$
a_n = \left(1 + \frac{1}{n} \right)^n
$$

$$
\lim_{n \to \infty} \left(1 + \frac{1}{n} \right)^n = e
$$

Thus, the limit is:

$$
e \approx 2.718
