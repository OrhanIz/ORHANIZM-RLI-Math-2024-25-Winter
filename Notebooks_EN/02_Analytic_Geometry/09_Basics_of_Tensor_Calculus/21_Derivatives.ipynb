## 1. Compute the Derivatives

### **Basic Derivatives:**
#### **(a)**
$$
y(x) = -3x + 3
$$
Derivative:
$$
y'(x) = -3
$$

#### **(b)**
$$
y(x) = \pi x + \sin(1)
$$
Derivative:
$$
y'(x) = \pi
$$

#### **(c)**
$$
y(x) = 4 + \sin(2)
$$
Derivative:
$$
y'(x) = 0
$$

#### **(d)**
$$
y(x) = 2x^3 - 3x^2 + 8x - 9
$$
Derivative:
$$
y'(x) = 6x^2 - 6x + 8
$$

#### **(e)**
$$
y(x) = 6x^{1/3}
$$
Derivative:
$$
y'(x) = 6 \cdot \frac{1}{3} x^{-2/3} = 2 x^{-2/3} = \frac{2}{\sqrt[3]{x^2}}
$$

#### **(f)**
$$
y(x) = \sqrt{x} = x^{1/2}
$$
Derivative:
$$
y'(x) = \frac{1}{2} x^{-1/2} = \frac{1}{2\sqrt{x}}
$$

#### **(g)**
$$
y(x) = \cos(x) + \sin(x)
$$
Derivative:
$$
y'(x) = -\sin(x) + \cos(x)
$$

#### **(h)**
$$
y(x) = 2\sin(x)\cos(x)
$$
Using product-to-sum formula:
$$
y(x) = \sin(2x)
$$
Derivative:
$$
y'(x) = 2\cos(2x)
$$

#### **(i)**
$$
y(x) = x\sin(x)
$$
Using product rule:
$$
y'(x) = x\cos(x) + \sin(x)
$$

#### **(j)**
$$
y(x) = (x+1)(x+1) = (x+1)^2
$$
Derivative:
$$
y'(x) = 2(x+1)
$$

#### **(k)**
$$
y(x) = \frac{x}{x+1}
$$
Using quotient rule:
$$
y'(x) = \frac{(x+1) \cdot 1 - x \cdot 1}{(x+1)^2} = \frac{1}{(x+1)^2}
$$

#### **(l)**
$$
y(x) = (x+1)e^x
$$
Using product rule:
$$
y'(x) = (x+1)e^x + e^x = (x+2)e^x
$$

#### **(m)**
$$
y(x) = \sin(x^2)
$$
Using chain rule:
$$
y'(x) = \cos(x^2) \cdot 2x = 2x \cos(x^2)
$$

#### **(n)**
$$
y(x) = e^{-2x}
$$
Using chain rule:
$$
y'(x) = e^{-2x} \cdot (-2) = -2e^{-2x}
$$

#### **(o)**
$$
y(x) = \frac{1}{\sin(x+1)}
$$
Using chain and quotient rules:
$$
y'(x) = -\frac{\cos(x+1)}{\sin^2(x+1)}
$$

#### **(p)**
$$
y(x) = \sqrt{2x+1} = (2x+1)^{1/2}
$$
Using chain rule:
$$
y'(x) = \frac{1}{2}(2x+1)^{-1/2} \cdot 2 = \frac{1}{\sqrt{2x+1}}
$$

## 2. Prove:
### **Derivative of \( \ln(\sin(x)) \)**
#### **To Prove:**
$$
\frac{d}{dx} (\ln(\sin(x))) = \cot(x)
$$

#### **Solution:**
Using the chain rule:
$$
\frac{d}{dx} \ln(\sin(x)) = \frac{1}{\sin(x)} \cdot \cos(x)
$$

Since \( \frac{\cos(x)}{\sin(x)} = \cot(x) \), we get:
$$
\frac{d}{dx} \ln(\sin(x)) = \cot(x)
$$

Thus, proved.

## 3. Verify:
### **For \( f(x) = \cos(x) \), verify \( f''(x) = -f(x) \)**
#### **Solution:**
First derivative:
$$
f'(x) = -\sin(x)
$$

Second derivative:
$$
f''(x) = -\cos(x)
$$

Since \( -\cos(x) = -f(x) \), we confirm:
$$
f''(x) = -f(x)
$$

Thus, verified.

## 4. L'H么pital's Rule

### **First Limit:**
$$
\lim_{x\to 0} \frac{\sin{x}}{x}
$$

$$

### **Second Limit:**

\lim_{x\to \infty} \frac{\ln x}{x}

### **Third Limit:**
$$
\lim_{x\to \infty} \frac{e^x}{x}

\lim_{x\to \infty} \frac{e^x}{x} = \infty

## 5. Physics: Velocity and Acceleration

Given position function:
$$
x(t) = 3t^2 - 6t + 1
$$

### **Velocity:**
$$
V(t) = x'(t) = 6t - 6
$$

At \( t = 2 \):
$$
V(2) = 6(2) - 6 = 6
$$

### **Acceleration:**
$$
a(t) = V'(t) = x''(t) = 6
$$

At \( t = 2 \):
$$
a(2) = 6
$$

Thus, at \( t = 2 \):

- Velocity: \( V(2) = 6 \)
- Acceleration: \( a(2) = 6 \)
