# **Solutions to Integral Problems**

## **1. Compute the Indefinite Integrals**

### **(a) Compute \( \int 1 \,dx \)**

#### **What is Given?**
We need to integrate the constant function \( 1 \).

#### **What Are We Doing?**
We use the basic integral formula:

$$
\int c \,dx = cx + C
$$

where \( C \) is the constant of integration.

#### **Solution:**
Applying the formula:

$$
\int 1 \,dx = x + C
$$

### **(b) Compute \( \int (x^2 +2) \,dx \)**

#### **What is Given?**
We need to integrate the polynomial function \( x^2 + 2 \).

#### **What Are We Doing?**
We integrate each term separately using the **power rule**:

$$
\int x^n \,dx = \frac{x^{n+1}}{n+1} + C
$$

#### **Solution:**
Applying the rule:

$$
\int x^2 \,dx + \int 2 \,dx
$$

$$
\frac{x^3}{3} + 2x + C
$$

---

### **(c) Compute \( \int 2\sin(x) \,dx \)**

#### **What is Given?**
We need to integrate the function \( 2\sin(x) \).

#### **What Are We Doing?**
Using the standard integral identity:

$$
\int \sin(x) \,dx = -\cos(x)
$$

#### **Solution:**
Applying the identity:

$$
\int 2\sin(x) \,dx = -2\cos(x) + C
$$

---

### **(d) Compute \( \int \frac{3}{x} \,dx \)**

#### **What is Given?**
We need to integrate the function \( \frac{3}{x} \).

#### **What Are We Doing?**
Using the **logarithmic integration rule**:

$$
\int \frac{1}{x} \,dx = \ln|x|
$$

#### **Solution:**
Applying the formula:

$$
\int \frac{3}{x} \,dx = 3\ln|x| + C
$$

---

### **(e) Compute \( \int \frac{1}{x^2} \,dx \)**

#### **What is Given?**
We need to integrate the function \( \frac{1}{x^2} \).

#### **What Are We Doing?**
Rewriting as \( x^{-2} \) and applying the **power rule**:

$$
\int x^n \,dx = \frac{x^{n+1}}{n+1} + C
$$

#### **Solution:**
Since \( n = -2 \):

$$
\int x^{-2} \,dx = \frac{x^{-1}}{-1} = -\frac{1}{x} + C
$$`

## **2. Compute the Definite Integrals Over the Interval \( [0, \pi] \)**

### **(a) Compute \( \int_0^\pi (2x+1) \,dx \)**

#### **What is Given?**
We need to compute the definite integral of \( 2x+1 \) over \( [0, \pi] \).

#### **What Are We Doing?**
First, find the indefinite integral:

$$
\int (2x+1) \,dx = x^2 + x
$$

Then, evaluate it at \( x = \pi \) and \( x = 0 \).

#### **Solution:**
Evaluating:

$$
\left[ x^2 + x \right]_0^\pi = (\pi^2 + \pi) - (0)
$$

$$
= \pi^2 + \pi
$$

Thus, the area under the curve is:

$$
\pi^2 + \pi
$$

## **3. Finding the Bounded Area**

### **What is Given?**
Find the area enclosed by:
- \( x = 1 \), \( x = 2 \)
- \( y = 0 \) (x-axis)
- \( y = x^2 + 1 \) (upper curve)

### **What Are We Doing?**
We compute the definite integral:

$$
A = \int_1^2 (x^2 + 1) \,dx
$$

### **Solution:**
Integrating:

$$
\int (x^2 + 1) \,dx = \frac{x^3}{3} + x
$$

Evaluating:

$$
\left[ \frac{x^3}{3} + x \right]_1^2
$$

$$
= \left( \frac{8}{3} + 2 \right) - \left( \frac{1}{3} + 1 \right)
$$

$$
= \frac{10}{3}
$$

Thus, the area is:

$$
\frac{10}{3}
$$

## **4. Compute the Area Under the Sine Curve**

### **What is Given?**
Compute:

$$
\int_0^\pi \sin(x) \,dx
$$

### **What Are We Doing?**
Using the identity:

$$
\int \sin(x) \,dx = -\cos(x)
$$

### **Solution:**
Evaluating at the bounds:

$$
\left[ -\cos(x) \right]_0^\pi
$$

$$
= (-\cos(\pi)) - (-\cos(0))
$$

$$
= (-(-1)) - (-1) = 2
$$

Thus, the area under the sine curve is:

$$
2
$$

## **6. Distance Traveled by a Particle**

### **What is Given?**
The position function:

$$
x(t) = 3t^2 - 6t + 1
$$

### **What Are We Doing?**
1. Compute the velocity:
   
   $$
   V(t) = x'(t) = 6t - 6
   $$

2. Find when \( V(t) = 0 \):

   $$
   6t - 6 = 0 \Rightarrow t = 1
   $$

3. Compute total distance:

   $$
   \int_0^1 |6t - 6| \,dt + \int_1^2 |6t - 6| \,dt
   $$

Since \( V(t) < 0 \) for \( 0 \leq t < 1 \), take absolute values:

$$
\int_0^1 -(6t - 6) \,dt + \int_1^2 (6t - 6) \,dt
$$

Evaluating:

$$
= 3 + 3 = 6
$$

Thus, the total distance is:

$$
6
$$
