# **Solutions to Integral Problems**

## **1. Compute the Indefinite Integrals**

### **(a) Compute \( \int 1 \,dx \)**

#### **What is Given?**
We need to integrate the constant function \( 1 \).

#### **What Are We Doing?**
We use the basic integral formula:

$$
\int c \,dx = cx + C
$$

where \( C \) is the constant of integration.

#### **Solution:**
Applying the formula:

$$
\int 1 \,dx = x + C
$$

### **(b) Compute \( \int (x^2 +2) \,dx \)**

#### **What is Given?**
We need to integrate the polynomial function \( x^2 + 2 \).

#### **What Are We Doing?**
We integrate each term separately using the **power rule**:

$$
\int x^n \,dx = \frac{x^{n+1}}{n+1} + C
$$

#### **Solution:**
Applying the rule:

$$
\int x^2 \,dx + \int 2 \,dx
$$

$$
\frac{x^3}{3} + 2x + C
$$

### **(c) Compute \( \int 2\sin(x) \,dx \)**

#### **What is Given?**
We need to integrate the function \( 2\sin(x) \).

#### **What Are We Doing?**
Using the standard integral identity:

$$
\int \sin(x) \,dx = -\cos(x)
$$

#### **Solution:**
Applying the identity:

$$
\int 2\sin(x) \,dx = -2\cos(x) + C
$$

### **(d) Compute \( \int \frac{3}{x} \,dx \)**

#### **What is Given?**
We need to integrate the function \( \frac{3}{x} \).

#### **What Are We Doing?**
Using the **logarithmic integration rule**:

$$
\int \frac{1}{x} \,dx = \ln|x|
$$

#### **Solution:**
Applying the formula:

$$
\int \frac{3}{x} \,dx = 3\ln|x| + C
$$

### **(e) Compute \( \int \frac{1}{x^2} \,dx \)**

#### **What is Given?**
We need to integrate the function \( \frac{1}{x^2} \).

#### **What Are We Doing?**
Rewriting as \( x^{-2} \) and applying the **power rule**:

$$
\int x^n \,dx = \frac{x^{n+1}}{n+1} + C
$$

#### **Solution:**
Since \( n = -2 \):

$$
\int x^{-2} \,dx = \frac{x^{-1}}{-1} = -\frac{1}{x} + C
$$`

## **2. Compute the Definite Integrals Over the Interval \( [0, \pi] \)**

### **(a) Compute \( \int_0^\pi (2x+1) \,dx \)**

#### **What is Given?**
We need to compute the definite integral of \( 2x+1 \) over \( [0, \pi] \).

#### **What Are We Doing?**
First, find the indefinite integral:

$$
\int (2x+1) \,dx = x^2 + x
$$

Then, evaluate it at \( x = \pi \) and \( x = 0 \).

#### **Solution:**
Evaluating:

$$
\left[ x^2 + x \right]_0^\pi = (\pi^2 + \pi) - (0)
$$

$$
= \pi^2 + \pi
$$

Thus, the area under the curve is:

$$
\pi^2 + \pi
$$

## **3. Compute the Area of a Region Bounded by \( x=1 \), \( x=2 \), \( y=0 \), and \( y=x^2+1 \)**

The area is computed as:

$$
A = \int_1^2 (x^2 + 1) \, dx
$$

### **Step 1: Compute the Integral**
Using the integral formula:

$$
\int (x^2 + 1) \, dx = \frac{x^3}{3} + x + C
$$

Evaluating from \( 1 \) to \( 2 \):

$$
\left[ \frac{2^3}{3} + 2 \right] - \left[ \frac{1^3}{3} + 1 \right]
$$

$$
\left[ \frac{8}{3} + 2 \right] - \left[ \frac{1}{3} + 1 \right] = \frac{14}{3} - \frac{4}{3} = \frac{10}{3}
$$

✅ **Final Result:**
\[
A = \frac{10}{3}
\]

## **4. Compute the Area Under the Sine Curve Over \( [0, \pi] \)**

The area under the sine curve is given by:

$$
P = \int_0^\pi \sin(x) \, dx
$$

Using the integral formula:

$$
\int \sin(x) \, dx = -\cos(x) + C
$$

Evaluating from \( 0 \) to \( \pi \):

$$
\left[ -\cos(\pi) \right] - \left[ -\cos(0) \right]
$$

$$
\left[ -(-1) \right] - \left[ -(1) \right] = 1 + 1 = 2
$$

✅ **Final Result:**
\[
P = 2
\]

## **6. Compute the Distance Traveled by a Moving Particle**

Given the position function:

$$
x(t) = 3t^2 - 6t + 1
$$

### **Step 1: Compute the Velocity**
The velocity function is:

$$
V(t) = \frac{dx}{dt} = 6t - 6
$$

### **Step 2: Compute the Total Distance**
The total distance is:

$$
D = \int_0^2 |V(t)| \, dt
$$

Find where \( V(t) = 0 \):

$$
6t - 6 = 0 \Rightarrow t = 1
$$

Thus, we split the integral:

$$
D = \int_0^1 |6t - 6| \, dt + \int_1^2 |6t - 6| \, dt
$$

For \( 0 \leq t \leq 1 \), \( V(t) \) is negative, so:

$$
\int_0^1 (6 - 6t) \, dt
$$

For \( 1 \leq t \leq 2 \), \( V(t) \) is positive, so:

$$
\int_1^2 (6t - 6) \, dt
$$

Computing each:

$$
\int_0^1 (6 - 6t) \, dt = [6t - 3t^2]_0^1 = (6 - 3) - (0) = 3
$$

$$
\int_1^2 (6t - 6) \, dt = [3t^2 - 6t]_1^2 = (12 - 12) - (3 - 6) = 3
$$

Thus, the total distance traveled is:

$$
D = 3 + 3 = 6
$$

✅ **Final Result:**
\[
D = 6
\]
