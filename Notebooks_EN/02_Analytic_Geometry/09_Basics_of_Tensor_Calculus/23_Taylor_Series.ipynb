# Solutions to Taylor Series Problems

## 1. Finding the Taylor Series

### **(a) Taylor series of \( f(x) = \cos(x) \) around \( x = 0 \) up to the 4th degree**
The Taylor series expansion of \( \cos(x) \) at \( x = 0 \) is:

$$
\cos(x) = \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n)!} x^{2n}
$$

Expanding up to the 4th-degree term:

$$
\cos(x) \approx 1 - \frac{x^2}{2!} + \frac{x^4}{4!}
$$

$$
\cos(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24}
$$

### **(b) Taylor series of \( h(x) = \frac{1}{1-x} \) around \( x = 0 \) up to the 4th degree**
The Taylor series expansion of \( \frac{1}{1-x} \) at \( x = 0 \) is:

$$
\frac{1}{1-x} = \sum_{n=0}^{\infty} x^n
$$

Expanding up to the 4th-degree term:

$$
\frac{1}{1-x} \approx 1 + x + x^2 + x^3 + x^4
$$

### **(c) Taylor series of \( g(x) = \sin(x) \) around \( x = \pi \) up to the 4th degree**
The Taylor series expansion of \( \sin(x) \) around \( x = \pi \) is:

$$
\sin(x) = \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n+1)!} (x - \pi)^{2n+1}
$$

Expanding up to the 4th-degree term:

$$
\sin(x) \approx \sin(\pi) + \cos(\pi)(x - \pi) - \frac{\sin(\pi)}{2!} (x - \pi)^2 - \frac{\cos(\pi)}{3!} (x - \pi)^3
$$

Since \( \sin(\pi) = 0 \) and \( \cos(\pi) = -1 \), we get:

$$
\sin(x) \approx -(x - \pi) + \frac{(x - \pi)^3}{6}
$$

## 2. Finding the Tangent Line to \( f(x) = e^{\sin(x)} \) at \( x_0 = \pi \)

### **Step 1: Compute the derivative**
Given:

$$
f(x) = e^{\sin(x)}
$$

Differentiate using the chain rule:

$$
f'(x) = e^{\sin(x)} \cos(x)
$$

At \( x_0 = \pi \):

$$
f'(\pi) = e^{\sin(\pi)} \cos(\pi) = e^0 \cdot (-1) = -1
$$

The function value at \( x_0 = \pi \):

$$
f(\pi) = e^{\sin(\pi)} = e^0 = 1
$$

### **Step 2: Find the equation of the tangent line**
Using the formula:

$$
y = f'(x_0) (x - x_0) + f(x_0)
$$

Substituting values:

$$
y = -1 (x - \pi) + 1
$$

$$
y = 1 - (x - \pi)
$$

or:

$$
y = -x + \pi + 1
$$
