## 22. Extremum

6. The profit function is $P(u) = -2u^2 + 50u - 300$, where $u$ is the number of units sold. Find the number of units that maximize profit.

7. You have 10 meters of string, and you need to use it to enclose the largest possible rectangular. Find the dimensions of the rectangle.

8. Find extremum od $f(x) = x^2 + 3x - 5$.

9. Find extremum of $f(x) =\frac{x^2+2x+1}{x-1}$.

# Solutions to Extremum Problems

## 1. Maximizing Profit

### **Given function:**
$$
P(u) = -2u^2 + 50u - 300
$$

### **Finding the maximum profit:**
To find the number of units \( u \) that maximizes profit, we take the derivative:

$$
P'(u) = \frac{d}{du} (-2u^2 + 50u - 300)
$$

$$
P'(u) = -4u + 50
$$

Set \( P'(u) = 0 \) to find critical points:

$$
-4u + 50 = 0
$$

$$
u = \frac{50}{4} = 12.5
$$

Since the second derivative:

$$
P''(u) = -4 < 0
$$

is negative, the function has a **maximum** at \( u = 12.5 \). Since the number of units should be an integer, the closest values are \( u = 12 \) or \( u = 13 \).

Thus, the profit is maximized at approximately:

$$
u = 12 \text{ or } u = 13
$$

## 2. Maximizing the Area of a Rectangle with 10m of String

### **Step 1: Define the problem**
Let the length of the rectangle be \( x \) and the width be \( y \). The perimeter constraint is:

$$
2x + 2y = 10
$$

or:

$$
x + y = 5
$$

The area of the rectangle is:

$$
A = xy
$$

### **Step 2: Express in one variable**
From the perimeter equation:

$$
y = 5 - x
$$

Substituting into the area equation:

$$
A(x) = x(5 - x)
$$

or:

$$
A(x) = 5x - x^2
$$

### **Step 3: Find the maximum area**
Differentiate:

$$
A'(x) = 5 - 2x
$$

Set \( A'(x) = 0 \):

$$
5 - 2x = 0
$$

$$
x = \frac{5}{2} = 2.5
$$

From the second derivative:

$$
A''(x) = -2 < 0
$$

Since it is negative, \( A(x) \) has a maximum at \( x = 2.5 \).

Thus, the dimensions of the rectangle that enclose the largest area are:

$$
x = 2.5, \quad y = 2.5
$$

**The largest possible rectangle is a square with sides of** \( 2.5 \) meters.

## 3. Finding Extremum of \( f(x) = x^2 + 3x - 5 \)

### **Step 1: Take the derivative**
$$
f'(x) = 2x + 3
$$

### **Step 2: Find critical points**
Set \( f'(x) = 0 \):

$$
2x + 3 = 0
$$

$$
x = -\frac{3}{2}
$$

### **Step 3: Second derivative test**
$$
f''(x) = 2 > 0
$$

Since \( f''(x) > 0 \), the function has a **minimum** at \( x = -\frac{3}{2} \).

### **Step 4: Compute function value at critical point**
$$
f(-\frac{3}{2}) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5
$$

$$
= \frac{9}{4} - \frac{9}{2} - 5
$$

$$
= \frac{9}{4} - \frac{18}{4} - \frac{20}{4}
$$

$$
= \frac{9 - 18 - 20}{4} = \frac{-29}{4}
$$

Thus, the minimum value is:

$$
\left( -\frac{3}{2}, -\frac{29}{4} \right)
$$

## 4. Finding Extremum of \( f(x) = \frac{x^2 + 2x + 1}{x - 1} \)

### **Step 1: Take the derivative using quotient rule**
Let:

$$
f(x) = \frac{x^2 + 2x + 1}{x - 1}
$$

Using quotient rule:

$$
f'(x) = \frac{(2x + 2)(x - 1) - (x^2 + 2x + 1)(1)}{(x - 1)^2}
$$

Expanding the numerator:

$$
(2x + 2)(x - 1) = 2x^2 - 2x + 2x - 2 = 2x^2 - 2
$$

$$
(2x^2 - 2) - (x^2 + 2x + 1) = 2x^2 - 2 - x^2 - 2x - 1
$$

$$
= x^2 - 2x - 3
$$

Thus:

$$
f'(x) = \frac{x^2 - 2x - 3}{(x - 1)^2}
$$

### **Step 2: Find critical points**
Set \( f'(x) = 0 \):

$$
x^2 - 2x - 3 = 0
$$

Factorizing:

$$
(x - 3)(x + 1) = 0
$$

Solutions:

$$
x = 3, \quad x = -1
$$

### **Step 3: Second derivative test**
Compute second derivative:

$$
f''(x) = \frac{(2x - 2)(x - 1)^2 - (x^2 - 2x - 3)(2(x - 1))}{(x - 1)^4}
$$

Evaluating at \( x = 3 \):

$$
f''(3) = \frac{2(3) - 2}{(3 - 1)^2} = \frac{4}{4} = 1 > 0
$$

Since \( f''(3) > 0 \), \( x = 3 \) is a **local minimum**.

Evaluating at \( x = -1 \):

$$
f''(-1) = \frac{2(-1) - 2}{(-1 - 1)^2} = \frac{-4}{4} = -1 < 0
$$

Since \( f''(-1) < 0 \), \( x = -1 \) is a **local maximum**.

Thus, the function has:
- **Local maximum** at \( x = -1 \)
- **Local minimum** at \( x = 3 \)
