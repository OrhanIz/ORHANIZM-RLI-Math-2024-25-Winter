## 25. Differential Equations

1. Solve the following first-order ordinary differential equations:
   - $y'(x)= y(x)$
   - $y'(x) = \frac{1}{2y(x)}$
  
3. Solve the first-order ordinary differential equations using the method of separation of variables for $x$ and $y=y(x)$:

   - $\frac{dy}{dx} = \frac{x}{y}$
   - $\frac{dy}{dx} = \frac{y}{x}$
   - $\frac{dy}{dx} = xy$

4. Solve the second-order ordinary differential equations:

   * $y''(x) + y'(x) = 0$, with boundary conditions $y(0) = 2$ and $y'(0) = -1$

   * $y''(x) - y(x)= 0$, with boundary conditions $y(0) = 2$ and $y'(0) = 0$

   * $\frac{d^2 y(x)}{dx^2} = -\omega^2 y(x)$.
  
5. Check if the function $\psi(t, x) = A \cos(\omega t + kx)$ is a solution of the second-order partial differential equation (the so-called "wave equation"), where $v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda}$:

$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

# **Solutions to Differential Equations**

## **1. Solving First-Order Ordinary Differential Equations**
We solve the given differential equations step by step.

### **(a) Solve \( y'(x) = y(x) \)**

#### **Step 1: Recognizing the Type of ODE**
This is a **first-order linear differential equation**:

$$
\frac{dy}{dx} = y
$$

#### **Step 2: Separating Variables**
Rewriting:

$$
\frac{dy}{y} = dx
$$

#### **Step 3: Integrating Both Sides**
$$
\int \frac{1}{y} dy = \int dx
$$

$$
\ln |y| = x + C
$$

#### **Step 4: Exponentiate Both Sides**
$$
y = e^{x + C} = C_1 e^x
$$

✅ **Final Solution:**
$$
y(x) = C e^x
$$

---

### **(b) Solve \( y'(x) = \frac{1}{2y(x)} \)**

#### **Step 1: Separating Variables**
$$
2y dy = dx
$$

#### **Step 2: Integrating Both Sides**
$$
\int 2y dy = \int dx
$$

$$
y^2 = x + C
$$

#### **Step 3: Solve for \( y \)**
$$
y = \pm \sqrt{x + C}
$$

✅ **Final Solution:**
$$
y(x) = \pm \sqrt{x + C}
$$

---

## **2. Solving First-Order ODEs Using Separation of Variables**
We solve the given equations using separation of variables.

### **(a) Solve \( \frac{dy}{dx} = \frac{x}{y} \)**

#### **Step 1: Separating Variables**
$$
y dy = x dx
$$

#### **Step 2: Integrating Both Sides**
$$
\int y dy = \int x dx
$$

$$
\frac{y^2}{2} = \frac{x^2}{2} + C
$$

#### **Step 3: Solve for \( y \)**
$$
y^2 = x^2 + C
$$

$$
y = \pm \sqrt{x^2 + C}
$$

✅ **Final Solution:**
$$
y(x) = \pm \sqrt{x^2 + C}
$$

---

### **(b) Solve \( \frac{dy}{dx} = \frac{y}{x} \)**

#### **Step 1: Separating Variables**
$$
\frac{dy}{y} = \frac{dx}{x}
$$

#### **Step 2: Integrating Both Sides**
$$
\int \frac{dy}{y} = \int \frac{dx}{x}
$$

$$
\ln |y| = \ln |x| + C
$$

#### **Step 3: Exponentiate Both Sides**
$$
y = C x
$$

✅ **Final Solution:**
$$
y(x) = Cx
$$

---

### **(c) Solve \( \frac{dy}{dx} = xy \)**

#### **Step 1: Separating Variables**
$$
\frac{dy}{y} = x dx
$$

#### **Step 2: Integrating Both Sides**
$$
\int \frac{dy}{y} = \int x dx
$$

$$
\ln |y| = \frac{x^2}{2} + C
$$

#### **Step 3: Exponentiate Both Sides**
$$
y = e^C e^{x^2/2}
$$

$$
y = C e^{x^2/2}
$$

✅ **Final Solution:**
$$
y(x) = C e^{x^2/2}
$$

---

## **3. Solving Second-Order Ordinary Differential Equations**

### **(a) Solve \( y''(x) + y'(x) = 0 \) with \( y(0) = 2 \) and \( y'(0) = -1 \)**

#### **Step 1: Write the Characteristic Equation**
$$
r^2 + r = 0
$$

#### **Step 2: Solve for \( r \)**
$$
r(r+1) = 0 \Rightarrow r = 0, -1
$$

#### **Step 3: General Solution**
$$
y(x) = C_1 + C_2 e^{-x}
$$

#### **Step 4: Apply Initial Conditions**
$$
y(0) = C_1 + C_2 = 2
$$

$$
y'(x) = -C_2 e^{-x}
$$

$$
y'(0) = -C_2 = -1 \Rightarrow C_2 = 1
$$

$$
C_1 = 1
$$

✅ **Final Solution:**
$$
y(x) = 1 + e^{-x}
$$

---

### **(b) Solve \( y''(x) - y(x) = 0 \) with \( y(0) = 2 \) and \( y'(0) = 0 \)**

#### **Step 1: Write the Characteristic Equation**
$$
r^2 - 1 = 0
$$

#### **Step 2: Solve for \( r \)**
$$
r = \pm 1
$$

#### **Step 3: General Solution**
$$
y(x) = C_1 e^x + C_2 e^{-x}
$$

#### **Step 4: Apply Initial Conditions**
$$
y(0) = C_1 + C_2 = 2
$$

$$
y'(x) = C_1 e^x - C_2 e^{-x}
$$

$$
y'(0) = C_1 - C_2 = 0
$$

Solving for \( C_1 \) and \( C_2 \):

$$
C_1 = C_2 = 1
$$

✅ **Final Solution:**
$$
y(x) = e^x + e^{-x}
$$

---

### **(c) Solve \( y''(x) = -\omega^2 y(x) \)**

#### **Step 1: Characteristic Equation**
$$
r^2 + \omega^2 = 0
$$

#### **Step 2: Solve for \( r \)**
$$
r = \pm i\omega
$$

#### **Step 3: General Solution**
$$
y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x)
$$

✅ **Final Solution:**
$$
y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x)
$$

---

## **4. Verify the Wave Equation Solution**
We need to check if:

$$
\psi(t, x) = A \cos(\omega t + kx)
$$

is a solution to:

$$
\frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0
$$

### **Step 1: Compute Second Derivatives**
$$
\frac{\partial^2 \psi}{\partial t^2} = -\omega^2 A \cos(\omega t + kx)
$$

$$
\frac{\partial^2 \psi}{\partial x^2} = -k^2 A \cos(\omega t + kx)
$$

Since \( v^2 = \frac{\omega^2}{k^2} \), we see that:

$$
\frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0
$$

✅ **Thus, \( \psi(t, x) \) satisfies the wave equation.**
