## 25. Differential Equations

1. Solve the following first-order ordinary differential equations:
   - $y'(x)= y(x)$
   - $y'(x) = \frac{1}{2y(x)}$
  
3. Solve the first-order ordinary differential equations using the method of separation of variables for $x$ and $y=y(x)$:

   - $\frac{dy}{dx} = \frac{x}{y}$
   - $\frac{dy}{dx} = \frac{y}{x}$
   - $\frac{dy}{dx} = xy$

4. Solve the second-order ordinary differential equations:

   * $y''(x) + y'(x) = 0$, with boundary conditions $y(0) = 2$ and $y'(0) = -1$

   * $y''(x) - y(x)= 0$, with boundary conditions $y(0) = 2$ and $y'(0) = 0$

   * $\frac{d^2 y(x)}{dx^2} = -\omega^2 y(x)$.
  
5. Check if the function $\psi(t, x) = A \cos(\omega t + kx)$ is a solution of the second-order partial differential equation (the so-called "wave equation"), where $v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda}$:

$$
\frac{\partial^2 \psi(t, x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t, x)}{\partial x^2} = 0.
$$

# **Solutions to Differential Equations**

## **1. Solve First-Order Ordinary Differential Equations**

### **(a) Solve \( y'(x) = y(x) \)**

#### **What is Given?**
We need to solve the first-order ordinary differential equation:

$$
y'(x) = y(x)
$$

#### **What Are We Doing?**
This is a **separable differential equation**, which can be rewritten as:

$$
\frac{dy}{dx} = y
$$

Rearrange terms:

$$
\frac{dy}{y} = dx
$$

#### **Solution:**
Integrating both sides:

$$
\int \frac{dy}{y} = \int dx
$$

Using the integral formula \( \int \frac{1}{y} dy = \ln|y| \):

$$
\ln |y| = x + C
$$

Exponentiating both sides:

$$
y = e^{x + C} = e^C e^x
$$

Since \( e^C \) is an arbitrary constant, we define it as \( C_1 \):

$$
y(x) = C_1 e^x
$$

---

### **(b) Solve \( y'(x) = \frac{1}{2y(x)} \)**

#### **What is Given?**
We need to solve:

$$
y'(x) = \frac{1}{2y(x)}
$$

#### **What Are We Doing?**
Rearrange to separate variables:

$$
2y \, dy = dx
$$

#### **Solution:**
Integrating both sides:

$$
\int 2y \, dy = \int dx
$$

Using the formula \( \int y \, dy = \frac{y^2}{2} \):

$$
y^2 = x + C
$$

Taking the square root:

$$
y(x) = \pm\sqrt{x + C}
$$`

---

## **2. Solve Differential Equations Using Separation of Variables**

### **(a) Solve \( \frac{dy}{dx} = \frac{x}{y} \)**

#### **What is Given?**
Rewriting:

$$
y \, dy = x \, dx
$$

#### **Solution:**
Integrating both sides:

$$
\int y \, dy = \int x \, dx
$$

$$
\frac{y^2}{2} = \frac{x^2}{2} + C
$$

Multiplying by 2:

$$
y^2 = x^2 + C
$$

Taking the square root:

$$
y = \pm\sqrt{x^2 + C}
$$

---

### **(b) Solve \( \frac{dy}{dx} = \frac{y}{x} \)**

#### **What is Given?**
Rearrange:

$$
\frac{dy}{y} = \frac{dx}{x}
$$

#### **Solution:**
Integrating both sides:

$$
\ln |y| = \ln |x| + C
$$

Exponentiating:

$$
y = Cx
$$

---

### **(c) Solve \( \frac{dy}{dx} = xy \)**

#### **What is Given?**
Rearrange:

$$
\frac{dy}{y} = x \, dx
$$

#### **Solution:**
Integrating both sides:

$$
\int \frac{dy}{y} = \int x \, dx
$$

$$
\ln |y| = \frac{x^2}{2} + C
$$

Exponentiating:

$$
y = C e^{x^2 / 2}
$$

---

## **3. Solve Second-Order Differential Equations**

### **(a) Solve \( y''(x) + y'(x) = 0 \) with \( y(0) = 2 \), \( y'(0) = -1 \)**

#### **What is Given?**
The characteristic equation:

$$
r^2 + r = 0
$$

Factorizing:

$$
r(r + 1) = 0 \Rightarrow r = 0, -1
$$

#### **Solution:**
The general solution is:

$$
y(x) = C_1 + C_2 e^{-x}
$$

Applying boundary conditions:

1. \( y(0) = 2 \) gives \( C_1 + C_2 = 2 \).
2. \( y'(x) = -C_2 e^{-x} \), so \( y'(0) = -C_2 = -1 \).

Solving:

$$
C_2 = 1, \quad C_1 = 1
$$

Final solution:

$$
y(x) = 1 + e^{-x}
$$

---

### **(b) Solve \( y''(x) - y(x) = 0 \) with \( y(0) = 2 \), \( y'(0) = 0 \)**

#### **What is Given?**
The characteristic equation:

$$
r^2 - 1 = 0
$$

Solving:

$$
r = \pm 1
$$

#### **Solution:**
The general solution:

$$
y(x) = C_1 e^x + C_2 e^{-x}
$$

Using initial conditions:

1. \( y(0) = 2 \Rightarrow C_1 + C_2 = 2 \)
2. \( y'(x) = C_1 e^x - C_2 e^{-x} \), so \( y'(0) = C_1 - C_2 = 0 \)

Solving:

$$
C_1 = C_2 = 1
$$

Final solution:

$$
y(x) = e^x + e^{-x}
$$

---

### **(c) Solve \( y''(x) = -\omega^2 y(x) \)**

#### **What is Given?**
The characteristic equation:

$$
r^2 + \omega^2 = 0
$$

Solving:

$$
r = \pm i\omega
$$

#### **Solution:**
The general solution:

$$
y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x)
$$

---

## **4. Check if \( \psi(t, x) = A \cos(\omega t + kx) \) Satisfies the Wave Equation**

#### **What is Given?**
The **wave equation**:

$$
\frac{\partial^2 \psi}{\partial t^2} - v^2 \frac{\partial^2 \psi}{\partial x^2} = 0
$$

where:

$$
v = \frac{\omega}{k}
$$

#### **What Are We Doing?**
We compute second derivatives:

1. **First time derivative:**
   $$
   \frac{\partial \psi}{\partial t} = -A \omega \sin(\omega t + kx)
   $$
2. **Second time derivative:**
   $$
   \frac{\partial^2 \psi}{\partial t^2} = -A \omega^2 \cos(\omega t + kx)
   $$

3. **First spatial derivative:**
   $$
   \frac{\partial \psi}{\partial x} = -A k \sin(\omega t + kx)
   $$
4. **Second spatial derivative:**
   $$
   \frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx)
   $$

Substituting into the wave equation:

$$
(-A \omega^2 \cos(\omega t + kx)) - v^2 (-A k^2 \cos(\omega t + kx)) = 0
$$

Since \( v^2 = \frac{\omega^2}{k^2} \), the equation holds, proving \( \psi(t, x) \) is a solution.
