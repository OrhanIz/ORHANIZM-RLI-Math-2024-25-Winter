## 20. Limits of Real Functions

1. Compute:
   - $\displaystyle\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}$

2. Find:
   
   - $\displaystyle \lim_{x \to 0} \frac{\sin(3x)}{2x+1}$.

4. Find the asymptotes of the function:
  
   - $f(x) = \frac{x^2 - 1}{x^2 + 1}$
   - $g(x) = \frac{\sin(x)}{x^2+1}$


# Solutions to Limits of Real Functions

## 1. Computing Limits

### **First Limit:**
$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}
$$

#### **Solution:**
- Divide both numerator and denominator by \( x^4 \):

$$
\frac{x^3 + 2x^2}{x^4 - 3x^3} = \frac{\frac{x^3}{x^4} + \frac{2x^2}{x^4}}{\frac{x^4}{x^4} - \frac{3x^3}{x^4}}
$$

- Simplify each term:

$$
\frac{\frac{1}{x} + \frac{2}{x^2}}{1 - \frac{3}{x}}
$$

- As \( x \to \infty \), the terms \( \frac{1}{x} \) and \( \frac{2}{x^2} \) approach 0, so:

$$
\lim_{x \to \infty} \frac{0 + 0}{1 - 0} = 0
$$

Thus, the limit is:

$$
0
$$

---

### **Second Limit:**
$$
\lim_{x \to 0} \frac{\sin(3x)}{2x+1}
$$

#### **Solution:**
- Substitute \( x = 0 \):

$$
\frac{\sin(3(0))}{2(0) + 1} = \frac{\sin(0)}{1} = \frac{0}{1} = 0
$$

Thus, the limit is:

$$
0
$$

---

## 2. Finding Asymptotes

### **For** \( f(x) = \frac{x^2 - 1}{x^2 + 1} \)

#### **Horizontal Asymptotes:**
- Compute the limit at infinity:

$$
\lim_{x \to \infty} \frac{x^2 - 1}{x^2 + 1}
$$

- Divide numerator and denominator by \( x^2 \):

$$
\frac{1 - \frac{1}{x^2}}{1 + \frac{1}{x^2}}
$$

- As \( x \to \infty \), \( \frac{1}{x^2} \to 0 \), so:

$$
\frac{1 - 0}{1 + 0} = 1
$$

Thus, the horizontal asymptote is:

$$
y = 1
$$

#### **Vertical Asymptotes:**
- Since the denominator \( x^2 + 1 \) never equals 0, there are no vertical asymptotes.

---

### **For** \( g(x) = \frac{\sin(x)}{x^2+1} \)

#### **Horizontal Asymptotes:**
- Compute the limit at infinity:

$$
\lim_{x \to \infty} \frac{\sin(x)}{x^2+1}
$$

- Since \( -1 \leq \sin(x) \leq 1 \), we bound the function:

$$
-\frac{1}{x^2+1} \leq \frac{\sin(x)}{x^2+1} \leq \frac{1}{x^2+1}
$$

- Since \( \frac{1}{x^2+1} \to 0 \) as \( x \to \infty \), by the **Squeeze Theorem**:

$$
\lim_{x \to \infty} g(x) = 0
$$

Thus, the horizontal asymptote is:

$$
y = 0
$$

#### **Vertical Asymptotes:**
- Since the denominator \( x^2 + 1 \) never equals 0, there are no vertical asymptotes.
