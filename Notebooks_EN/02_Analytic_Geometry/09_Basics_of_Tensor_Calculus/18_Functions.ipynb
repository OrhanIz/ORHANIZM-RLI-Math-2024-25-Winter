# Solutions to Function Problems

## 1. Drawing Functions and Evaluating at \( x = 2 \)

### Given functions:

$$
f(x) = x^2, \quad g(x) = \sqrt{x}, \quad h(x) = \frac{1}{x}, \quad j(x) = \sin(x)
$$

### Values at \( x = 2 \):

$$
f(2) = 2^2 = 4
$$

$$
g(2) = \sqrt{2} \approx 1.414
$$

$$
h(2) = \frac{1}{2} = 0.5
$$

$$
j(2) = \sin(2) \approx 0.909
$$

### Visualization:
To draw these functions in Geogebra:
- Open Geogebra and use the graphing interface.
- Enter the functions: \( f(x) \), \( g(x) \), \( h(x) \), and \( j(x) \).
- Add the points \( (2, f(2)) \), \( (2, g(2)) \), \( (2, h(2)) \), and \( (2, j(2)) \).

## 2. Function Compositions

### Given functions:

$$
f(x) = 3x - 1, \quad g(x) = \sqrt{x}
$$

### Function compositions:

$$
f(g(x)) = f(\sqrt{x}) = 3\sqrt{x} - 1
$$

$$
g(f(x)) = g(3x - 1) = \sqrt{3x - 1}
$$

$$
f(f(x)) = f(3x - 1) = 3(3x - 1) - 1 = 9x - 4
$$

$$
g(g(x)) = g(\sqrt{x}) = \sqrt{\sqrt{x}} = x^{\frac{1}{4}}
$$

### Visualization:
In Geogebra:
- Enter \( f(g(x)) = 3\sqrt{x} - 1 \), \( g(f(x)) = \sqrt{3x - 1} \), \( f(f(x)) = 9x - 4 \), and \( g(g(x)) = x^{1/4} \).
- Use different colors for each function.

---

## 3. Exponential and Logarithmic Functions

### Given:

$$
f(x) = e^x, \quad g(x) = \ln(x)
$$

### Checking compositions:

$$
f(g(x)) = e^{\ln(x)} = x
$$

$$
g(f(x)) = \ln(e^x) = x
$$

### Observation:
Since \( f(g(x)) = x \) and \( g(f(x)) = x \), these functions are inverses of each other.

---

## 4. Finding Inverse Function

### Given function:

$$
f = \{(1,7), (2,9), (3,11)\}
$$

### Inverse function:

$$
f^{-1} = \{(7,1), (9,2), (11,3)\}
$$

### Observation:
The function is one-to-one and onto, so its inverse is well-defined.

---

## 5. Non-Invertible Function

### Given function:

$$
f = \{(1,7), (2,7), (3,11)\}
$$

### Inverse function:
**Not defined** because the function is not one-to-one.

### Explanation:
Since \( f(1) = 7 \) and \( f(2) = 7 \), there is no unique inverse mapping.

---

## 6. Finding the Inverse Function of \( f(x) = x - 1 \)

### Given:

$$
f(x) = x - 1
$$

### To find the inverse:

$$
y = x - 1
$$

Solve for \( x \):

$$
x = y + 1
$$

Thus, the inverse function is:

$$
f^{-1}(x) = x + 1
$$

### Visualization:
In Geogebra:
- Enter \( f(x) = x - 1 \) and \( f^{-1}(x) = x + 1 \).
- Plot the identity line \( y = x \) to show the symmetry.
